```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from functions import weather_generator, count_anomalies
```

```python
# Use the weather_generator function to generate weather
# data for switzerland
mean = 0.0
sigma = 0.94
n = 100000

# Generate weather data
weather = weather_generator(mean, sigma, n)
```

```python
# For our data array we count the number of values that are
# greater than 5.1
count = count_anomalies(weather, 5.1)
print(count)
```

```python
# Edit the mean to increase by 1.25C
new_mean = mean + 1.25

# Generate new weather data
new_weather = weather_generator(new_mean, sigma, n)

# Count the number of anomalies
new_count = count_anomalies(new_weather, 5.1)
print(new_count)
```

```python
# Plot probability density function
# for the original weather data and the new weather data
plt.hist(weather, bins=100, density=True, label='Original', alpha=0.5)
# plot the mean as a vertical line
plt.axvline(x=mean, color='blue', linestyle='--', label='Mean')
plt.hist(new_weather, bins=100, density=True, label='Shifted mean', alpha=0.5)
# plot the mean as a vertical line
plt.axvline(x=new_mean, color='orange', linestyle='--', label='New mean')
# Plot the threshold as a vertical line
plt.axvline(x=5.1, color='red', linestyle='--', label='Threshold')
plt.xlabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Explore the implications of changing the mean and sigma
new_sigma = sigma + 0.5
weather_new_sigma = weather_generator(mean, new_sigma, n)

# Count the number of values greater than 5.1
count_new_sigma = count_anomalies(weather_new_sigma, 5.1)
print(count_new_sigma)

# Change both the mean and sigma
weather_new_mean_sigma = weather_generator(new_mean, new_sigma, n)

# Count the number of values greater than 5.1
count_new_mean_sigma = count_anomalies(weather_new_mean_sigma, 5.1)
print(count_new_mean_sigma)
```

```python
# Plot the probability density function for the new sigma
plt.hist(weather_new_sigma, bins=100, density=True, label='New sigma', alpha=0.5)
# plot the mean as a vertical line
plt.axvline(x=mean, color='blue', linestyle='--', label='Mean')
plt.hist(weather_new_mean_sigma, bins=100, density=True, label='New mean and sigma', alpha=0.5)
# plot the mean as a vertical line
plt.axvline(x=new_mean, color='orange', linestyle='--', label='New mean')
# plot the threshold as a vertical line
plt.axvline(x=5.1, color='red', linestyle='--', label='Threshold')
plt.xlabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Between 2070-2100 with a business as usual scenario
# the mean temperature is expected to increase by 4.6C
# and the standard deviation is expected to double
weather_2070_2100 = weather_generator(4.6, 2*sigma, n)

# Plot the probability density function for the new sigma
plt.hist(weather_2070_2100, bins=100, density=True, label='2070-2100', alpha=0.5)
# plot the mean as a vertical line
plt.axvline(x=np.mean(weather_2070_2100), color='blue', linestyle='--', label='Mean')
# plot the threshold as a vertical line
plt.axvline(x=5.1, color='red', linestyle='--', label='Threshold')
plt.xlabel('Temperature (C)')

# Count the number of values greater than 5.1
count_2070_2100 = count_anomalies(weather_2070_2100, 5.1)
print(count_2070_2100)
```
